```python
from datetime import datetime
from GameBoard import GameBoard
from Random_Agent import RandomAgent
from Agent import Agent
from ExpectimaxAgent import ExpectimaxAgent
```

```python
def check_win(board: GameBoard):
    return board.get_max_tile() >= 2048


int_to_string = ['UP', 'DOWN', 'LEFT', 'RIGHT']
```

```python
agent: Agent
board: GameBoard
agent = ExpectimaxAgent()
board = GameBoard()
done = False
moves = 0
board.render()
start = datetime.now()
while not done:
    action = agent.play(board)
    print('Next Action: "{}"'.format(
        int_to_string[action]), ',   Move: {}'.format(moves))
    done = board.play(action)
    done = done or check_win(board)
    board.render()
    moves += 1

print('\nTotal time: {}'.format(datetime.now() - start))
print('\nTotal Moves: {}'.format(moves))
if check_win(board):
    print("WON THE GAME!!!!!!!!")
else:
    print("BOOOOOOOOOO!!!!!!!!!")
```

El siguiente cdigo permite correr el agente con los distintas heursticas que permite, y una variacin en los parmetros max_depth entre 1 y 4:

```python
iterations = 7
heurs = ["sum", "empty", "value", "smoothness"]
for h in heurs:
    for i in range (1, 5):
        print("Heuristic: {}".format(h), ", Depth: {}".format(i))
        avg_time, avg_moves, wins, loses = 0, 0, 0, 0

        for _ in range(iterations):
            agent: Agent
            board: GameBoard
            agent = ExpectimaxAgent(i, h)
            board = GameBoard()
            done = False
            moves = 0
            start = datetime.now()
            while not done:
                action = agent.play(board)
                done = board.play(action)
                done = done or check_win(board)
                moves += 1

            avg_time += (datetime.now() - start).total_seconds()
            avg_moves += moves
            if check_win(board):
                wins += 1
            else:
                loses += 1
        
        avg_time /= iterations
        avg_moves /= iterations
        print('Average Time (s): {}'.format(avg_time))
        print('Average Moves: {}'.format(avg_moves))
        print('Wins: {}'.format(wins))
        print('Loses: {}'.format(loses))
```
